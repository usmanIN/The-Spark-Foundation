# Coronavirus Pandemic (COVID-19) India

```python
import pandas as pd
import matplotlib as mtp
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
```

```python
india = df.iloc[:-1,1:]
```

```python
india = india.set_index('state_name',drop=True)
```

# Cases By State

```python
india.plot.bar(stacked=True,figsize=(20,10))
```

```python
india.sort_values(by='positive',ascending=False).style.bar(subset=['positive'],color=['#5fba7d']).bar(subset=['death'],color=['#d65f5f']).set_table_styles([{'selector' : '','props' : [('font-size','10pt'),('border','1px solid black')]}]) 
```

```python
india['positive'].corr(india['death'])
```
